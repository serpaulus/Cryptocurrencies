```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import hvplot.pandas
```

```python
#Load processes Iris data set
file_path = "Resources/new_iris_data.csv"
df_iris = pd.read_csv(file_path)
df_iris.head()
```

```python
#Standardize the data
iris_scaled = StandardScaler().fit_transform(df_iris)
iris_scaled
```

```python
# Initialize PCA model
pca = PCA(n_components=2)
```

```python
# Get two principal components for the iris data.
iris_pca = pca.fit_transform(iris_scaled)
```

```python
# Transform the iris data to two principal components.
df_iris_pca = pd.DataFrame(data=iris_pca, columns=['PC1', 'PC2'])
df_iris_pca.head()
```

```python
import plotly.figure_factory as ff
```

```python
# Create the dendrogram
fig = ff.create_dendrogram(df_iris_pca, color_threshold=0)
fig.update_layout(width=800, height=500)
fig.show()
```

```python
agg = AgglomerativeClustering(n_clusters=3)
model = agg.fit(df_iris_pca)
```

```python
# Add a new class column to the dataframe
df_iris_pca['class'] = model.labels_
df_iris_pca.head()
```

```python
# Create a scatter plot of the two principal components
df_iris_pca.hvplot.scatter(
    x='PC1', 
    y='PC2',
    hover_cols=['class'],
     by='class')
```
